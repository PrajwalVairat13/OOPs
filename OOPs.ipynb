1. What is Object-Oriented Programming (OOP) ?
---
- Object-Oriented Programming (OOP) is a programming paradigm that organizes software design around data, or objects, rather than functions and logic, focusing on objects with unique attributes and behaviors.

2. What is a class in OOP ?
---
- In Object-Oriented Programming (OOP), a class is a blueprint or template for creating objects.

3. What is an object in OOP ?
---
- An object can be defined as a data field that has unique attributes and behavior.

4. what is the difference between abstraction and encapsulation ?
---
- Abstraction focuses on hiding implementation details and showcasing only essential functionalities, while encapsulation bundles data and methods into a single unit and controls access to them.

5. What are dunder methods in Python ?
---
- Python dunder methods are the special predefined methods in Python having two prefixes and two suffix underscores in the method name

6. Explain the concept of inheritance in OOP.
---
- Inheritance in object-oriented programming (OOP) is a mechanism where a new class (derived or subclass) inherits properties and behaviors (methods) from an existing class (base or superclass).

7. What is polymorphism in OOP ?
---
- In Object-Oriented Programming (OOP), polymorphism, meaning "many forms," allows objects of different classes to be treated as objects of a common superclass, enabling them to respond to the same method call in different ways.

8. How is encapsulation achieved in Python ?
---
- In Python, encapsulation, a core object-oriented programming concept, is achieved by bundling data (attributes) and methods that operate on that data within a class, and then controlling access to these attributes and methods using access modifiers (public, protected, and private)

9. What is a constructor in Python ?
---
- Constructors in Python is a special class method for creating and initializing an object instance at that class.

10. What are class and static methods in Python ?
---
- In Python, class methods are functions tied to the class itself, not individual objects, and they typically take the class as the first argument (cls).
- Static methods, on the other hand, are functions associated with a class but don't have any special access to the class or its instances

11.  What is method overloading in Python ?
---
- Method overloading is a feature of object-oriented programming where a class can have multiple methods with the same name but different parameters.

12. What is method overriding in OOP ?
---
- In Object-Oriented Programming (OOP), method overriding is a mechanism where a subclass provides its own specific implementation for a method that is already defined in its superclass.

13. What is a property decorator in Python ?
---
- The Python property decorator makes attributes in a class act like read-only properties.

14. Why is polymorphism important in OOP ?
---
- Polymorphism is crucial in OOP because it allows you to treat objects of different classes as objects of a common type, enabling code reuse, flexibility, and extensibility.

15. what is an abstract class in python ?
---
- An abstract class in Python is typically created to declare a set of methods that must be created in any child class built on top of this abstract class.

16. What are the advantages of OOP ?
---
- Object-Oriented Programming (OOP) offers several advantages, including enhanced code organization, reusability, and flexibility.

17. What is the difference between a class variable and an instance variable ?
---
- Difference between a class variable (also known as a static variable) and an instance variable is their scope and how they are associated with objects of a class.

18. What is multiple inheritance in Python ?
---
- In Python, multiple inheritance allows a class (the child class) to inherit attributes and methods from multiple parent classes.

19. Explain the purpose of '__str__' and '__repr__'  methods in Python ?
---
- The __str__ method returns a readable description of an object for users, while __repr__ gives a detailed or technical string for developers and debugging.


20. What is the significance of the ‘super()’ function in Python ?
---
- The super() function in Python is crucial for implementing inheritance and facilitating code reuse within class hierarchies.

21. What is the significance of the __del__ method in Python ?
---
- In Python, the __del__() method is a destructor that's automatically called by the garbage collector when an object's reference count reaches zero, signaling that the object is about to be destroyed.

22. What is the difference between @staticmethod and @classmethod in Python ?
---
- '@staticmethod' doesn't access the class or instance, while '@classmethod' gets the class as a parameter and can modify class-level data.

23. How does polymorphism work in Python with inheritance ?
---
- Polymorphism in Python with inheritance allows different subclasses to define their own versions of a method, and the correct version is called automatically based on the object's class, even when using a common parent reference.

24. What is method chaining in Python OOP ?
---
- Method chaining in Python allows you to call multiple methods on the same object in a single line of code, improving readability and conciseness.

25. What is the purpose of the __call__ method in Python ?
---
- The purpose of the __call__ method in Python is to make instances of a class callable, allowing them to be invoked like functions.
